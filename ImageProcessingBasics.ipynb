#Converting RGB to Grayscale using OpenCV

```python
import cv2
```

```python
img = cv2.imread('/content/cristiano-ronaldo-portugal-world-cup-wallpaper-preview.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
Grayscale = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(Grayscale)
```

```python
Grayscale.shape
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(Grayscale)
```

#saving the grayscale image


```python
cv2.imwrite('ronaldo_grayscale.jpg', Grayscale)
```
